### Filling Missin Values Automatically

```python
# Get small subset of NFL Dataset
subset_Nf1_data = nf1_data.loc[:,'EPA':'Season'].head()
subset_Nf1_data
```

```python
# We can use pandas fillna() function to fill missing values
subset_Nf1_data.fillna(0)
```

```python
#replace all NA's the values that came directly after in it the same column,
#then replace all missing na's with 0 
subset_Nf1_data.fillna(method = 'bfill',axis = 0).fillna("0")
```

```python
# Filling missing value is also known as imputation.
```

```python
sf_permits.fillna(method = 'bfill',axis = 0).fillna('0')
```
